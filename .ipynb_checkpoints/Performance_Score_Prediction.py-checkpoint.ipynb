```python
def prediction(x,thetas):   # For All Samples at one by one.
    y_=thetas[0]+thetas[1]*x
    return y_
```

```python
def gradient(x,y,thetas):
    m=x.shape[0]
    grad=np.zeros((2,))
    for k in range(m):
        val=x[k]
        y_=prediction(val,thetas)
        y_act=y[k]
        grad[0]+=(y_ - y_act)
        grad[1]+=(y_ - y_act)*val
    return grad/m
```

```python
def Error(x,y,thetas):
    m=x.shape[0]
    total_error=0.0
    for j in range(m):
        y_=prediction(x[j],thetas)
        total_error += (y_ - y[j])**2
        
    return total_error
```

```python
def gradientDescent(x,y,learning_rate=0.1,max_steps=100):   # Here x and y are complete samples
    
    thetas=np.zeros((2,))
    theta_list=[]
    error_list=[]
    
    for i in range(max_steps):
        
        grad=gradient(x,y,thetas)
        thetas[0]=thetas[0]-learning_rate*grad[0]  # Update Thetas
        thetas[1]=thetas[1]-learning_rate*grad[1]
        e=Error(x,y,thetas)
        error_list.append(e)
        theta_list.append((thetas[0],thetas[1]))
        
    
    return thetas,theta_list,error_list
```

```python
thetas,theta_list,error_list=gradientDescent(X,Y)
```

```python
print(thetas)
theta_array=np.array(theta_list)
error_array=np.array(error_list)
```

```python
train_predictions=prediction(X,thetas)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,color='navy',alpha=0.5,s=80)
plt.xlabel('Time of Study',fontsize=14)
plt.ylabel('Performance Score',fontsize=14)
plt.title('Performance vs Time of Study',fontsize=16)
plt.plot(X,train_predictions,color='orange',label='Best Fit Line')
plt.legend()
plt.grid()
plt.style.use('seaborn')
plt.show()
```

```python
# Deriving using mathematically
def r2score(y,y_pred):
    num=np.sum((y-y_pred)**2)
    den=np.sum((y-y.mean())**2)
    score=1-num/den
    return score*100
```
